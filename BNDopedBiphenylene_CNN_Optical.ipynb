```python
import pandas as pd

# Defining the database directory:
data_path = "/content/drive/MyDrive/BNDopedBiphData/data/BNDopedBiphenylene_Optical.csv"
# Storing the dataset in a Pandas dataframe:
data_frame = pd.read_csv(data_path)
display(data_frame)
```

```python
import numpy as np

# Specifying the dataset size:
rows = 6
columns = 4

# Specifying x data:
x_data = []
for index, row in data_frame.iterrows():
    #print(index, '  ', row['Atom Arrangements'])
    atom_list = list(row['Atom Arrangements'].split())
    #print(atom_list)
    atom_float_list = [float(x) for x in atom_list]
    #print(atom_int_list)
    #atom_matrix = np.array(atom_float_list)
    atom_matrix = np.reshape(atom_float_list, (rows, columns))
    #print(atom_matrix)
    x_data.append(atom_matrix)
#x_data = np.stack(x_data, axis=0)
x_data = np.array(x_data)
print(x_data)
print(x_data.shape)

# Specifying x data:
y_data = []
for index, row in data_frame.iterrows():
    opt_list = list(row['Optical (Tangential)'].split())
    opt_float_list = [float(x) for x in opt_list]
    opt_matrix = np.array(opt_float_list)
    #print(opt_matrix)
    y_data.append(opt_matrix)

y_data = np.array(y_data)
print(y_data)
print(y_data.shape)
```

```python
import tensorflow as tf
from tensorflow import keras
from keras import layers, models

#Definning CNN model:
model = models.Sequential()
model.add(layers.Flatten())
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(32, activation='relu'))
model.add(layers.Dense(24, activation='relu'))
model.add(layers.Dense(1))

input_shape=(None, rows, columns, 1)
model.build(input_shape)

model.summary()

# Compiling the model:
model.compile(loss="mean_squared_error", optimizer="RMSProp")
```

```python
# Importing MinMaxScaler from sklearn to normalize the data:
from sklearn.preprocessing import MinMaxScaler

# Creating the scaler:
y_scaler = MinMaxScaler()

# Defining variables:
batch_size = (len(data_frame) - 1) #number of training samples in one forward/backward pass
num_epoch = 10 #one epoch means one forward and one backward pass of all the training samples

for i in range(5):
  y_array = np.reshape(y_data[i], (-1, 1)) #Reshaping 1D array to 2D array
  y_array = y_scaler.fit_transform(y_array)
  y_data[i] = np.reshape(y_array,-1)

print(y_data)
```

```python
actual_y = []

x_test = np.array(x_data[0])
x_test = np.reshape(x_test, (1, rows, columns)) #Reshaping to 2D array
y_test = np.array(y_data[0])
y_test = np.reshape(y_test, (1, -1)) #Reshaping to 2D array
actual_y.append(y_data[0])

print("x test shape: ", x_test.shape)
print("y test shape: ", y_test.shape)
print("x test: ", x_test)
print("y test: ", y_test)
print("----------------------------")
# Train:
x_train = np.delete(x_data, 0, axis=0)
y_train = np.delete(y_data, 0, axis=0)
print("x train shape: ", x_train.shape)
print("y train shape: ", y_train.shape)
print("x train: ", x_train)
print("y train: ", y_train)
```

```python
predicted_y = []

# Model Training:
model_log = model.fit(x_train, y_train,
                    batch_size=batch_size,
                    epochs=num_epoch,
                    verbose=0,
                    validation_data=(x_test, y_test))

# Evaluating the model:
model.evaluate(x_test, y_test, verbose=0)

# Predicting the test data:
y_pred = model.predict(x_test)
predicted_y.append(y_pred)

actual_y = np.reshape(actual_y, (-1))
predicted_y = np.reshape(predicted_y, (-1))
print("Actual y values: ", actual_y)
print("Predicted y values: ", predicted_y)
```
